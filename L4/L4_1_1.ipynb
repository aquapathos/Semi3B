# 一般の jupyter notebook での画像表示方法
## 方法１ matplotlib

```
import matplotlib.pyplot as plt
% matplotlib inline
import cv2
file = "prof.jpg"
img = cv2.imread(file,-1)
plt.axis('off')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_RGB2BGR))
```
