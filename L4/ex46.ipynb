# jupyter noteboo 用コード

```
import cv2
import matplotlib.pyplot as plt
def main():
    file = "prof.jpg"
    img = cv2.imread(file,cv2.IMREAD_GRAYSCALE)
    print(img.shape)
    plt.axis('off')
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_GRAY2RGB))
                

main()
```

# 通常 python 用

```
import cv2
 
def main():
    file = "prof.jpg"
    img = cv2.imread(file,cv2.IMREAD_GRAYSCALE)
    cv2.imshow("prof",img)
    cv2.waitKey(0)
    cv2.destroyWindow("prof")
 
if __name__ == '__main__':
    main()
```
