```
import cv2
import numpy as np
# jupyter notebook でない場合は下の1行はコメントアウト
%matplotlib inline
 
fname = "noise.jpg" # 入力ファイル名
src_img = cv2.imread(fname)

dst_img = [np.array(src_img.shape) for i in range(4)]
# dst_img[]は src_img と同じサイズの画像配列４つ分の配列

dst_img[0] = cv2.blur(src_img,(5,5))
dst_img[1] = cv2.GaussianBlur(src_img,(5,5),1.0)
dst_img[2] = cv2.medianBlur(src_img,5)
dst_img[3] = cv2.bilateralFilter(src_img,5,75,75)

#cv2.imshow("Original", src_img)
#cv2.imshow("Blur",dst_img[0])
#cv2.imshow("Gaussian",dst_img[1])
#cv2.imshow("Median",dst_img[2])
#cv2.imshow("Bilateral",dst_img[3])

#cv2.waitKey(0)
#cv2.destroyAllWindows()

# jupyter notebook でない場合は以下はコメントアウト 
from matplotlib import pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 15,6
def plotimg(pos,label,img):
    plt.subplot(pos)
    plt.axis("off")
    plt.title(label)
    plt.imshow(img)
    
plotimg(151,"Original",src_img)
plotimg(152,"Blur",dst_img[0])
plotimg(153,"Gaussian",dst_img[1])
plotimg(154,"Median",dst_img[2])
plotimg(155,"Bilateral",dst_img[3])
```
